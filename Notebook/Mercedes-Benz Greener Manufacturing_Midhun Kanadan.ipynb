```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
from warnings import filterwarnings
filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# importing data set
train = pd.read_csv("train.csv")
```

```python
train.head()
```

#### If for any column(s), the variance is equal to zero, then you need to remove those variable(s).

```python
pd.options.display.float_format = '{:,.4f}'.format
variance = train.var()
variance = variance.reset_index()
variance.columns = ["id","values"]
variance= variance.sort_values("values",ascending=1)
variance
```

```python
# Our aim is to eliminate the variables that have a variance of 0 or a variance that is considered low, specifically less than 0.2.
var = variance.loc[variance["values"] < 0.2,"id"]
train_new = train.drop(var,axis=1)
train_new.drop("ID",axis=1,inplace=True)
train_new.head()
```

### Check for null and unique values for test and train sets.

```python
# Missing values
train_new[train_new.isnull().any(axis=1)]
```

#We don't have any missing values in the data

```python
c = train_new.corr().abs()
s = c.unstack()
s = pd.DataFrame(s)
s.reset_index(inplace = True)
s.head()
```

```python
s["flag"] = np.where(s["level_0"] == s["level_1"],"same","not same")
s.columns.values[2] = "corr"
s.head()
```

```python
# Remove the variables with correlation more than .9
name = s.loc[(s["corr"] > .9) & (s["flag"] != "same") ,"level_1"]
final_name = name.unique()
```

```python
train_new_updated = train_new.drop(final_name,axis=1)
train_new_updated.head()
```

```python
train_new_updated.info()
```

## Create dummy variables

```python
char = train_new_updated.select_dtypes(exclude='number')
num = train_new_updated.select_dtypes(include='number')
```

```python
char1 = pd.get_dummies(char.astype(str),drop_first=True)
char1.head()
```

```python
data3 = pd.concat([char1,num],axis=1)
data3.head()
```

## Run the XGBoost model and calculate the mape

```python
from sklearn.model_selection import train_test_split
import xgboost as xg
```

```python
#segregating ind and dep var
X = data3.drop("y",axis=1)
y = data3.loc[:,"y"]
```

```python
# Splitting
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, random_state = 185)
```

```python
# Instantiation
xgb_r = xg.XGBRegressor(objective ='reg:squarederror',n_estimators = 10, seed = 123)
```

```python
# Fitting the model
xgb_r.fit(X_train, y_train)
```

```python
# Predict the model
y_pred = xgb_r.predict(X_test)
```

```python
###### mape
final_predictions = pd.DataFrame()
final_predictions["y_test"] = y_test
final_predictions["y_pred"] = y_pred
final_predictions["mp"] = abs((final_predictions["y_test"]- final_predictions["y_pred"])/final_predictions["y_test"])
final_predictions.to_csv('predictions.csv', index=False)
(final_predictions.mp.mean())*100#mape
```
